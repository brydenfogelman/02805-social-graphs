* 
    * **Compare the lexical diversity scores for humor and romance fiction in 1.1. Which genre is more lexically diverse?**

```python
from __future__ import division

# CAUTION: Our lexical diversity here is the one defined in the 1st edition of the book, even though it should be the other way round
def lexical_diversity(text):
    return len(text) / len(set(text))

print "Lexical diversity for the humor fictions in the Brown Corpus: %f" % lexical_diversity(nltk.corpus.brown.words(categories='humor'))
print "Lexical diversity for the romance fictions in the Brown Corpus: %f" % lexical_diversity(nltk.corpus.brown.words(categories='romance'))
```

As we can see, the lexical diversity score is lower for the humor fiction, meaning we have an higher lexically diverse vocabulary in the humor fictions than in the romance fictions.

* 
    * **Define a function percent(word, text) that calculates how often a given word occurs in a text, and expresses the result as a percentage.**

```python
def percent(word, text):
    count = 0
    # For each word in the text, if the given word is in the text, add to the count 
    for a_word in text:
        if word == a_word:
            count += 1
    # Compute the percentage
    return 100 * (count / len(text))

print percent('a', text7)
```

    * 17) **Write a function that finds the 50 most frequently occurring words of a text that are not stopwords.**

```python
def top_50_words(text, language='english'):
    # Get the stopwords for the given language
    stopwords = nltk.corpus.stopwords.words(language)
    
    # Clean words : keep only those that are alphanumeric and convert them to lowercase
    words_cleaned = [w.lower() for w in text if w.isalpha() and w.lower() not in stopwords]
    
    fdist = nltk.FreqDist(words_cleaned)
    
    return sorted(fdist.items(), key=lambda (k,v): -v)[:50]

print top_50_words(text2)
```

* 
    * 18) **Write a program to print the 50 most frequent bigrams (pairs of adjacent words) of a text, omitting bigrams that contain stopwords.**

```python
def top_50_bigrams(text, language='english'):
    # Get the stopwords for the given language
    stopwords = nltk.corpus.stopwords.words(language)
    
    # Get bigrams for the given text, and keep only those that are actual words
    bigrams = [(w1.lower(), w2.lower()) for (w1, w2) in nltk.bigrams(text) 
               if w1.isalpha() and w2.isalpha()]
    # Now filter out any bigram containing a stopword
    bigrams = [(word1, word2) for (word1, word2) in bigrams 
               if word1 not in stopwords and word2 not in stopwords]
    
    fdist = nltk.FreqDist(bigrams)
    
    return sorted(fdist.items(), key=lambda (k,v): -v)[:50]

print top_50_bigrams(text1)
```

* **Work through exercise 2.8.23 on Zipf's law.**
    * Zipf's Law: Let f(w) be the frequency of a word w in free text. Suppose that all the words of a text are ranked according to their frequency, with the most frequent word first. Zipf's law states that the frequency of a word type is inversely proportional to its rank (i.e. f  r = k, for some constant k). For example, the 50th most common word type should occur three times as frequently as the 150th most common word type. 
        * **Write a function to process a large text and plot word frequency against word rank using pylab.plot. Do you confirm Zipf's law? (Hint: it helps to use a logarithmic scale). What is going on at the extreme ends of the plotted line?**

```python
def compute_words_frequency(text):
    # Clean words : keep only those that are alphanumeric and convert them to lowercase
    words_cleaned = [w.lower() for w in text if w.isalpha()]
    
    # Compute the words frequency
    fdist = nltk.FreqDist(words_cleaned)
    
    # Sort the word frequencies by decreasing count
    return sorted(fdist.items(), key=lambda (k,v): -v)

def plot_word_frequency(words_frequency):
    # Only keep the count 
    words_frequency = [count for (word, count) in words_frequency]

    # Generate the rankings 
    ranks = [x for x in range(1, len(words_frequency) + 1)]

    # Plot the word frequency against the ranks 
    plt.loglog(words_frequency, ranks)
    plt.show()

# Compute the words frequency
words_frequency = compute_words_frequency(text2)
plot_word_frequency(words_frequency)
```
