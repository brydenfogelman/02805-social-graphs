A connection on Twitter doesn't represent a mutual acceptance between two person. Whereas you become "friends" on Facebook, on Twitter you "follow" someone to keep updated on their activities.
```python
# Focusing on the hashtag DeadlineDay
q = '#DeadlineDay' 

count = 100

# Search for all tweets related to the hashtag #DeadlineDay
search_results = twitter_api.search.tweets(q=q, count=count)

# Retrieve statuses from the search 
statuses = search_results['statuses']

# Iterate through 5 more batches of results by following the cursor
for _ in range(5):
    print "Length of statuses", len(statuses)
    try:
        next_results = search_results['search_metadata']['next_results']
    except KeyError, e: # No more results when next_results doesn't exist
        break
        
    # Create a dictionary from next_results, which has the following form:
    # ?max_id=313519052523986943&q=NCAA&include_entities=1
    kwargs = dict([ kv.split('=') for kv in next_results[1:].split("&") ])
    
    search_results = twitter_api.search.tweets(**kwargs)
    statuses += search_results['statuses']

# Show one sample search result by slicing the list...
first_status = statuses[0]
print json.dumps(first_status, indent=1)
```

```python
# Extract the text from the first tweet 
first_status_text = first_status['text']

# Print the first Tweet 
print "First retrieved Tweet:"
print first_status_text
```

### Find the total retweet count for #DeadlineDay

```python
# Initalize the total retweet count at 0
rt_count = 0

# For each status, add to the total count the status' retweet count 
for status in statuses:
    rt_count += status['retweet_count']
    
print "Total number of retweet for %s : %d" % (q, rt_count)
```

### Find the tweet with the maximum number of retweets in the ones we retrieved

```python
# Take the first tweet as the maximum (save the status and the associated count)
max_rt_count_status = first_status
max_rt_count = first_status['retweet_count']

# Loop through each status and if the retweet count is greater or equal than the current maximum, update it 
for status in statuses[1:]:
    current_rt_count = status['retweet_count']
    if(current_rt_count >= max_rt_count):
        max_rt_count = current_rt_count
        max_rt_count_status = status

print "Tweet with the most retweets (%d):" % max_rt_count
print max_rt_count_status['text']
```
