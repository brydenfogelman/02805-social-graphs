```python
import scipy as sp

def show_adj_mat(D, DG):
    A_G = nx.adjacency_matrix(G)
    A_DG = nx.adjacency_matrix(DG)

    # adjacency matrices
    print(A_G.todense())
    print(A_DG.todense())
    
def show_list(D, DG):
    # linked lists
    print G.edges()
    print DG.edges()
    
show_adj_mat(G, DG)
show_list(G, DG)
```

From a linked list representation is can be a lot harder to tell which node or set of nodes is disconnected. In an adjancency matrix it is very easy to tell what is disconnected.

```python
import numpy as np
from numpy import linalg

A_G = nx.adjacency_matrix(G)
A_DG = nx.adjacency_matrix(DG)

A_G = linalg.matrix_power(A_G.todense(), 3)
print A_G[0,2]

A_DG = linalg.matrix_power(A_DG.todense(), 3)
print A_DG[0,2]
```

Q4: Questions 2.5 from NS

```python
B = nx.Graph()
B.add_nodes_from([1,2,3,4,5,6], bipartite=0)
B.add_nodes_from([7,8,9,10,11], bipartite=1)
edgelist = [(1,7),(2,9),(3,7),(3,8),(3,9),(4,9),(4,10),(5,9),(5,11),(6,11)]
B.add_edges_from(edgelist)

# nx.draw(B, with_labels=True)
# plt.show()
```

```python
A_B = nx.adjacency_matrix(B)
A_B.todense()
```

```python
from networkx.algorithms import bipartite

# top is green, bottom is purple
purple_nodes, green_nodes = bipartite.sets(B)

# finding the projects of green and purple onto the combined graph B
B_proj_purple = bipartite.projected_graph(B, purple_nodes)
B_proj_green = bipartite.projected_graph(B, green_nodes)

Adj_B_proj_purple = nx.adjacency_matrix(B_proj_purple)
print(Adj_B_proj_purple.todense())
Adj_B_proj_green = nx.adjacency_matrix(B_proj_green)
print(Adj_B_proj_green.todense())
```

```python
avg_deg_B = 2.*len(B.edges())/len(B.nodes())
print avg_deg_B

avg_deg_purple = 2.*len(B_proj_purple.edges())/len(B_proj_purple.nodes())
avg_deg_green = 2.*len(B_proj_green.edges())/len(B_proj_green.nodes())
print avg_deg_purple
print avg_deg_green
```

Is it surprising that the values are different from those obtained in point (c)?

No, in the projected graphs the connections are created that normally would not be found inside the orginal bipartite network. By doing this, we are artificially creating more neighbors for nodes in the projected graph in a sense.

Q2: Binning Integers

```python
# again, create a poisson distributed list that has 5000 elements
rand_nums = []
for ii in range(0,5000):
    rand_nums.append(np.random.poisson(10))

max_num = max(rand_nums)
min_num = min(rand_nums)
print max_num
print min_num
```

```python
# create a bin for every integer between and including the min and max number
bins = [num for num in range(min_num,max_num+1)]
hist, bin_edges = np.histogram(rand_nums, bins=bins)
print hist
print bin_edges
plt.bar(bin_edges[:-1], hist)
plt.show()
```

What is the sum over bin counts? Explain how the binning-vectors first and last element relates to the min and max from the Poisson distribution.

The sum over the bin counts is equal to 5000, the total sum of random numbers. The first and last bin correspond to the occurence of the min and max number respectively, the binning vector signifies what the value is.

Q3: Binning and plotting degree distributions

```python
# <k> = p(N-1) where <k> is the average degree, N is the total and p is an ER parameter
N = 5000
k = 10
p = k/(N-1.)

G = nx.erdos_renyi_graph(N, p)

nx.draw(G, node_size=25, node_color="#F7A8EE", node_shape='o', edge_color='.1', with_labels=False, width=1)
plt.show()
```

```python
degree_sequence=sorted(nx.degree(G).values())

# similar to before, find the max and min and createa bin vector containing all the numbers between them
max_num = max(degree_sequence)
min_num = min(degree_sequence)
bins = [num for num in range(min_num,max_num+1)]

# create a histogram with degree_sequence and the bins vector created above
hist, bins = np.histogram(degree_sequence, bins=bins)
plt.bar(bins[:-1], hist)
plt.title("Degree distribution plot of ER network")
plt.ylabel("# of occurences")
plt.xlabel("degree")
plt.show()
```

Q2: Paths

```python
k = 1.5
N = 200
p = 1.5/(200-1)

G = nx.erdos_renyi_graph(N, p)

nx.draw(G, node_size=25, node_color="#F7A8EE", node_shape='o', edge_color='.1', with_labels=False, width=0.8)
```

```python
gcc = max(nx.connected_component_subgraphs(G), key=len)
```

```python
rand_node = random.choice(gcc.nodes())
gcc.node[rand_node]
```

```python
paths = nx.single_source_shortest_path_length(gcc, rand_node)

close_nodes = []
for key in paths.keys():
    if paths[key] == 2:
       close_nodes.append(key) # print node that is 2 distances away from the rand_node
    
print close_nodes
```

```python
gcc_nodes = gcc.nodes()

nodes_color = ["#A0CBE2" for node in gcc_nodes]

for node in close_nodes:
    nodes_color[gcc_nodes.index(node)] = "#ff0000"
    
nodes_color[gcc_nodes.index(rand_node)] = "#000000"
    
nx.draw(gcc, node_size=25, node_color=nodes_color, node_shape='o', edge_color='.1', with_labels=False, width=0.8)
```

Plot the GCC with the following choice of colors. Starting node black ("#000000"). The nodes 2 steps away red ("#ff0000"). All other nodes blue ("#A0CBE2"). Again, I suggest using networkx.draw and reading the documentation carefully find out how to color individual nodes.
