* **Ex 16** Write a program to generate a table of lexical diversity scores (i.e. token/type ratios), as we saw in 1.1. Include the full set of Brown Corpus genres (nltk.corpus.brown.categories()). Which genre has the lowest diversity (greatest number of tokens per type)? Is this what you would have expected?

```python
# List of list used to store rows of table
tabs = []

# for each category in the Brown Corpus
for category in brown.categories():
    # extract the words
    words = brown.words(categories=category)
    # create a list of lists to contain the category, token, types and lexical diversity
    tabs.append([category, len(words), len(set(words)), len(words)/len(set(words))])
    
# use tabulate to print a nice table
print tabulate(tabs, headers=['Category','Tokens','Types','Lexical Diversity'])
```

* **Ex 17:** Write a function that finds the 50 most frequently occurring words of a text that are not stopwords.

```python
def freq_words(text):
    stopwords = nltk.corpus.stopwords.words('english')
    fdist = nltk.FreqDist(text)
    # find all words that are not stopwords
    content = [w for w in fdist.keys() if w.lower not in stopwords]
    # return 50 most frequent words
    return content[:50]

# test with text6
print freq_words(text6)
```

* **Ex 18:** Write a program to print the 50 most frequent bigrams (pairs of adjacent words) of a text, omitting bigrams that contain stopwords.

```python
from nltk.corpus import reuters

def freq_bigrams(text):
    # create set of stopwords
    stopwords = set(nltk.corpus.stopwords.words('english'))
    
    # find all pairs that are not in remove_words and create a new list of bigrams without those
    bigrams = [(pair[0].lower(), pair[1].lower()) for pair in nltk.bigrams(text) if all(c.lower() not in stopwords and c.isalpha() for c in pair)]
    # create the frequency distribution
    fdist = nltk.FreqDist(bigrams)
    # sort by value and get sort in descending order
    fdist_sorted = sorted(fdist.items(), key=operator.itemgetter(1), reverse=True)
    # get first 50
    print fdist_sorted[:50]

# test using brown
freq_bigrams(text1)
```
